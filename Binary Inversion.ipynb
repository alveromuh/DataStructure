```python
import sys
import os
try:
    sys.stdin = open('./input.txt', 'r')
    sys.stdout = open('./output.txt', 'w')
except:
    pass

# Membaca semua input yang diberikan 
for _ in range(int(input('Masukkan Banyaknya Test Case Yang Ingin Dilakukan : '))):
    print(" ")
    print("Tentukan Banyak dan Panjang dari Masing-Masing Binary String : ")
    print(" ")
    N,M = map(int,input().split())

    String_details = []

    # Membaca semua string yang diberikan dan menghitung jumlah yang ada dalam string biner
    # dan menyimpan detail dalam tuple
    print("Banyaknya String adalah", N)
    print("Masukkan string dengan Masing-Masing Memiliki Panjang", M)
    for _ in range(N):
        bin_string = input(str())  # input binary string 
        count_1 = bin_string.count('1')  # menghitung banyaknya angka 1 dalam string
        String_details.append((count_1,bin_string)) #memasangkan banyaknya angka 1 terhadap binary string
    
    # Menyortir string Biner dalam urutan menaik sehubungan dengan jumlah yang ada di string itu
    String_details.sort()
    
    print("Didapatkan String Setelah Diurutkan: ", String_details)

    Final_string = ""

    # Menghasilkan String biner terakhir
    for i in range(N):
        Final_string += String_details[i][1]
    print(" ")
    print("String Akhir Setelah Digabungkan: ",Final_string)
 
    ones_encounterted = 0
    min_inversions = 0

    # Traversing / Lintasi string Biner terakhir kami dihasilkan
    for i in range(N*M):
        if Final_string[i] == '1':
            # Menyimpan string biner yang mana terdapat karakter angka 1 dalam traversal
            ones_encounterted += 1
        else:
            # Jika 0 ditemukan, maka buat asumsi
            # Angka jumlah angka 1 yang ditemui sebelumnya dapat berpasangan dengan angka 0 saat ini
            # Tambahkan jumlah minimal inversi terhadap jumlah angka 1 yang ditemui
            min_inversions += ones_encounterted
    print(" ")
    print("Didapatkan Jumlah Minimum dari Binary Inversion : ",min_inversions)
```
